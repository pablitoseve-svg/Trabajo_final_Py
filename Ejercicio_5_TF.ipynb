# 5. Web Scrapping

Antes de nada, aseg√∫rate de abrir la carpeta 'Trabajo_Final', y ya ejecutado, instalar todas las librer√≠as

```python
import pandas as pd # Importamos las herramientas necesarias
from selenium import webdriver # El conductor del navegador
from selenium.webdriver.common.by import By # Para seleccionar elementos (ID, Class, Name...)
from selenium.webdriver.common.keys import Keys # Para pulsar teclas (Enter, Tab, etc.)
import time # Para hacer peque√±as pausas y ver qu√© ocurre

# Paso 1: Inicializar el navegador
# Esto abrir√° una ventana de Chrome controlada por un robot
driver = webdriver.Chrome()

# Configuraci√≥n recomendada: Espera impl√≠cita
# Si no encuentra un elemento, espera hasta 10 segundos antes de dar error.
# Esto es vital si la web es lenta cargando.
driver.implicitly_wait(10)

print("‚úÖ Navegador abierto y listo.")
```

```python
# Vamos a la p√°gina de Python
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
driver.get(url)

# Podemos verificar que estamos en el sitio correcto obteniendo el t√≠tulo
print(f"T√≠tulo de la p√°gina: {driver.title}")

# Hacemos una pausa de 2 segundos para que ve√°is la web abierta
time.sleep(2)
```

Aqu√≠, nos decargamos la tabla de datos

```python
html = driver.page_source
tables = pd.read_html(html)

df_companias = tables[0]
```

```python
['Symbol', 'Security', 'SEC filings', 'GICS Sector', 'GICS Sub-Industry',
 'Headquarters Location', 'Date first added', 'CIK', 'hideFounded']
```

Eliminamos las que no nos interesen para limpiarlo un poco'

```python
df_cambios = tables[1]
['Effective Date', 'Added', 'Removed', 'Reason']
```

Descargo los dos Dataframes (cambios y companias) y buscamos informaci√≥n que nos pueda interesar

```python

df_companias.to_csv("sp500_companias.csv", index=False)
df_cambios.to_csv("sp500_cambios.csv", index=False)
```

```python
print(df_companias.head())
print(df_cambios.head())
```

# Empresa m√°s antigua


```python
# Asegurar que Founded sea num√©rica
df_companias['hideFounded'] = pd.to_numeric(df_companias['hideFounded'], errors='coerce')

# Empresa m√°s antigua
empresa_mas_antigua = df_companias.loc[df_companias['hideFounded'].idxmin()]

print(empresa_mas_antigua)
```

# Empresas con m√°s salidas

```python
empresa_mas_salidas = (
    df_cambios['Removed']
    .dropna()
    .value_counts()
    .idxmax()
)

num_salidas = df_cambios['Removed'].value_counts().max()

print("Empresa con m√°s salidas del S&P 500:", empresa_mas_salidas)
print("N√∫mero de salidas:", num_salidas)
```

# Empresas con m√°s entradas

```python
empresa_mas_entradas = (
    df_cambios['Added']
    .dropna()
    .value_counts()
    .idxmax()
)

num_entradas = df_cambios['Added'].value_counts().max()

print("Empresa con m√°s entradas al S&P 500:", empresa_mas_entradas)
print("N√∫mero de entradas:", num_entradas)

```

Y finalmente, cerramos el navegador efectuando la pr√≥xima celda

```python
# Cerramos el navegador
driver.quit()
print("üëã Navegador cerrado correctamente.")
```
