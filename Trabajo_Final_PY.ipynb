# Golden Cross Analysis

```python
plt.figure(figsize=(10, 7))
plt.plot(san_OHLC_df_1['Close'], label='Close Price', color='blue')
plt.plot(san_OHLC_df_1['SMA_50'], label='50-Day SMA', color='orange')
plt.plot(san_OHLC_df_1['SMA_200'], label='200-Day SMA', color='green')
plt.title(f'Golden Cross Analysis for {TICKER_1}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
san_OHLC_df_1['SMA_50_prev'] = san_OHLC_df_1['SMA_50'].shift(1)
san_OHLC_df_1['SMA_200_prev'] = san_OHLC_df_1['SMA_200'].shift(1)

san_OHLC_df_1['crossover_signal'] = 0
san_OHLC_df_1.loc[(san_OHLC_df_1['SMA_50_prev'] < san_OHLC_df_1['SMA_200_prev']) & (san_OHLC_df_1['SMA_50'] > san_OHLC_df_1['SMA_200']), 'crossover_signal'] = 1
san_OHLC_df_1.loc[(san_OHLC_df_1['SMA_50_prev'] > san_OHLC_df_1['SMA_200_prev']) & (san_OHLC_df_1['SMA_50'] < san_OHLC_df_1['SMA_200']), 'crossover_signal'] = -1

crossover_points = san_OHLC_df_1[san_OHLC_df_1['crossover_signal'] != 0]
print(crossover_points[['Close', 'SMA_50', 'SMA_200', 'crossover_signal']])
```

```python
san_OHLC_df_1['crossover_signal'] = np.where(
    (san_OHLC_df_1['SMA_50'].shift(1) < san_OHLC_df_1['SMA_200'].shift(1)) & (san_OHLC_df_1['SMA_50'] > san_OHLC_df_1['SMA_200']), 1,
    np.where(
        (san_OHLC_df_1['SMA_50'].shift(1) > san_OHLC_df_1['SMA_200'].shift(1)) & (san_OHLC_df_1['SMA_50'] < san_OHLC_df_1['SMA_200']), -1,
        0
    )
)
crossover_points = san_OHLC_df_1[san_OHLC_df_1['crossover_signal'] != 0]

print(crossover_points[['Close', 'SMA_50', 'SMA_200', 'crossover_signal']])
```

```python
plt.figure(figsize=(10, 7))
plt.plot(san_OHLC_df_1['Close'], label='Close Price', color='blue')
plt.plot(san_OHLC_df_1['SMA_50'], label='50-Day SMA', color='orange')
plt.plot(san_OHLC_df_1['SMA_200'], label='200-Day SMA', color='green')

# Plot crossover points
plt.scatter(crossover_points[crossover_points['crossover_signal'] == 1].index,
            crossover_points[crossover_points['crossover_signal'] == 1]['SMA_50'],
            marker='^', color='green', s=100, label='Golden Cross (50-200)')
plt.scatter(crossover_points[crossover_points['crossover_signal'] == -1].index,
            crossover_points[crossover_points['crossover_signal'] == -1]['SMA_50'],
            marker='v', color='red', s=100, label='Death Cross (50-200)')

plt.title(f'Golden and Death Cross Analysis for {TICKER_1}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def get_column_by_ticker(df, ticker, data_type='Close'):
    """
    Devuelve la columna del DataFrame para un ticker dado y un tipo de dato (Close, Volume, etc.)
    Funciona con MultiIndex o columnas aplanadas.
    """
    # Caso 1: MultiIndex
    if isinstance(df.columns, pd.MultiIndex):
        if (data_type, ticker) in df.columns:
            return df[(data_type, ticker)]
        else:
            raise KeyError(f"No existe la columna ({data_type}, {ticker}) en el DataFrame.")

    # Caso 2: columnas aplanadas (tipo 'Close_BTC')
    flat_col = f"{data_type}_{ticker}"
    if flat_col in df.columns:
        return df[flat_col]

    raise KeyError(f"No existe la columna {flat_col} en el DataFrame.")
```
